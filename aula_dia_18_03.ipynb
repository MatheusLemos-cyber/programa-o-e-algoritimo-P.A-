```python
# Operadores aritimeticos

# Fazer umprograma que o usuario digita 02 numeros
# saida mostrar operações aritimeticas

#entrada

a = int(input('Digite o primeiro numero: '))
b = int(input('Digite o segundo numero: '))

#processamento

soma = a + b

#saida

print ('__'*31)
print ("____________________OPERADORES ARITIMETICOS___________________")
print ('__'*31)
print ("____________________________SOMA______________________________")
print (f" O resultado da SOMA de {a} mais {b} é igual a : {soma}")
```

```python
#entrada

a = int(input('Digite o primeiro numero: '))
b = int(input('Digite o segundo numero: '))

#processamento

soma = a + b
subtracao = a - b
multiplicao = a * b
divisao = a / b
div_int = a // b
resto = a % b
potenciacao = a ** b

#saida

print ('__'*30)
print ("")
print ("_____________soma_____________")
print (f" O resultado da SOMA de {a} mais {b} é igual a : {soma}")
print ("")
print ("___________subtração__________")
print (f" O resultado da subtracao de {a} menos {b} é igual a : {subtracao}")
print ("")
print ("___________multiplicação_________")
print (f" O resultado da multiplicação de {a} vezez {b} é igual a : {multiplicao}")
print ("")
print ("___________divisão_____________")
print (f" O resultado da divisão de {a} dividido por {b} é igual a: {divisao}")
print ("")
print ("___________divisão_____________")
print (f" O resultado da divisão de {a} dividido por {b} é igual a : {div_int}")
print ("")
print ("___________divisão_______________")
print (f" O resto da divisão de {a} dividido por {b} é igual a : {resto}")
print ("")
print ("___________potenciação_______________")
print (f" O resultado da potencia de {a} elevado a {b} é igual a : {potenciacao}")
```
