escrever um programa que faz conversão de temperatura

# ENTRADA
temperatura em Celsius

#PROCESSAMENTO
converter para Fahrenheit

#SAIDA
resultado.

#FORMULA
F= (C x 9/5) +32

```python
# entrada

celsius = float (input(f"digite a temperatura atual:"))

# processamento

fahrenheit = float  (celsius * 9/5) + 32

# saida

print (f"a temperatura em fahreheint de {celsius}° é: {fahrenheit}")
```
