```python
# ---------------------------------------------------------
# QUESTÃO 05 – Caixa Eletrônico (5 pontos)
# ---------------------------------------------------------

# Um caixa eletrônico possui notas de:

# R$ 100
# R$ 50
# R$ 20
# R$ 10
# R$ 5
# R$ 2
# R$ 1

# Solicite ao usuário o valor do saque e informe a menor
# quantidade possível de notas para realizar o pagamento.

# Exemplo:

# Digite o valor do saque: 287

# Saída esperada:

# 2 nota(s) de R$100
# 1 nota(s) de R$50
# 1 nota(s) de R$20
# 1 nota(s) de R$10
# 1 nota(s) de R$5
# 1 nota(s) de R$2

# Requisitos:
# - Utilizar laços de repetição.
# - Não utilizar listas.
# - Exibir apenas as notas utilizadas.

num1 = int (input("digite o valor que deseja sacar: "))
nota100 = 0
nota50 = 0
nota20 = 0
nota10 = 0
nota5 = 0
nota2 = 0
nota1 = 0
while num1 >= 0:

  if num1 >= 100:
    num1 -= 100
    nota100 += 1
    print(f"{nota100} nota(s) de 100")
  elif num1 >= 50:
    num1 -= 50
    nota50 += 1
    print(f"{nota50} nota(s) de 50")

  elif num1 >= 20:
    num1 -= 20
    nota20 += 1
    print(f"{nota20} nota(s) de 20")

  elif num1 >= 10:
    num1 -= 10
    nota10 += 1
    print(f"{nota10} nota(s) de 10")

  elif num1 >= 5:
    num1 -= 5
    nota5 += 1
    print(f"{nota5} nota(s) de 5")

  elif num1 >= 2:
    num1 -= 2
    nota2 += 1
    print(f"{nota2} nota(s) de 2")

  elif num1 >= 1:
    num1 -= 1
    nota1 += 1
    print(f"{nota1} nota(s) de 1")
```
