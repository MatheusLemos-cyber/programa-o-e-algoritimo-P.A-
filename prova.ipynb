```python
#exercicio 1 "par ou impar"

a = float (input("Digite o numero a descobrir se é par ou impar:"))

if a % 2 == 0:
  print(f"este numero é par: {a}")
else:
  print(f"este numero é impar: {a}")
```

```python
#exercicio 3 (Aprovação de aluno)

nota1 = float (input("Digite a nota do primeiro Bimestre: "))
if nota1 > 10:
  print(f"Digite um valor valido dentre 1 à 10.")
nota2 = float (input("Digite a nota do segundo Bimestre: "))
if nota1 > 10:
  print(f"Digite um valor valido dentre 1 à 10.")
nota3 = float (input("Digite a nota do terceiro Bimestre: "))
if nota1 > 10:
  print(f"Digite um valor valido dentre 1 à 10.")
nota4 = float (input("Digite a nota do quarto Bimestre: "))
if nota1 > 10:
  print(f"Digite um valor valido dentre 1 à 10.")

nota5 = (nota1 + nota2 + nota3 + nota4) / 4


if nota5 > 10:
  print(f"esta nota não é permitida, confira novamente se colocou as notas certas para não Ultrapassar o limite")

elif nota5 >= 7:
  print(f"o aluno foi aprovado com uma média de: {nota5}")

elif nota5 >= 5:
  print(f"o aluno está de recuperação com uma média de: {nota5}")

else:
  print(f"o aluno está reprovado com uma média de: {nota5}")
```

```python
#exercicio 4 (Classificação de idade)

idade = int (input("Qual sua idade? "))

if idade >= 60:
  print(f"Com esta idade você se classfica como um idoso: {idade}")

elif idade >= 18:
  print(f"desta idade você já é um adulto: {idade}")

elif idade >= 13:
  print(f"Parabéns você está na flor da idade, sendo um jovem de {idade}")

elif idade >= 0:
  print(f"Com está idade você ainda é uma criança {idade}")
```

```python
#exercicio 6 (Positivo, negativo ou zero)

numero = float (input("Digite um valor: "))

if numero == 0:
  print(f"este numero é 0.")

elif numero <= 0:
  print(f"este numero é negativo: {numero}")

else:
  print(f"este numero é positivo. {numero}")
```
