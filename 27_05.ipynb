```python
# Jogo Menu inteligente

# Criar um sistema que funcione até o usuário escolher sair.

# Menu

#1 -Par/ Impar
#2 -Positivo / Negativo
#3 Maior
#4 Sair

opcao = ""
num = ""

try:
  while opcao != 4:
    print("opção 1 - par/impar")
    print("opção 2 - positivo/negativo")
    print("opção 3 - maior")
    print("opção 4 - sair")
    print("-----------------------------------------------------------------")
    opcao = int(input("Digite a opção desejada: "))

    if opcao == 1:
      print("-------------------------------------------------------------------")
      num = float(input("Digite seu numero "))
      if num % 2 == 0:
        print("-------------------------------------------------------------------")
        print(f"seu numero escolhido {num} é par.")
      else:
        print("-------------------------------------------------------------------")
        print(f"seu numero escolhido {num} é impar")

    if opcao == 2:
      print("-----------------------------------------------------------------")
      num = float(input("Digite seu numero "))
      if num >= 0:
        print("-----------------------------------------------------------------")
        print(f"O numero {num} é um numero positivo")
      else:
        print("-----------------------------------------------------------------")
        print(f"o numero {num} é um numero negativo")

    elif opcao == 3:
      print("-----------------------------------------------------------------")
      num = float(input("Digite seu primeiro numero: "))
      num2 = float(input("Digite seu segundo numero: "))
      if num > num2:
        print("-----------------------------------------------------------------")
        print(f"O primeiro numero {num} é maior que o segundo numero {num2}")
      else:
        print("-----------------------------------------------------------------")
        print(f"O primeiro numero {num} é menor que o segundo numero {num2}")
    else:
      print("-----------------------------------------------------------------")
      print("Até uma proxima")
except:
  print("Digite somente numeros")
```
