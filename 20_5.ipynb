```python
# Escrever um programa que faz uma contagem regressiva de 10 até zero e imprimir
#uma menssagem de "acabou" no final

numero = 10
while numero >= 0:
  print(numero)
  numero -= 1
if numero == -1:
  print("ACABOUSSSEESESESEE")
```

```python
# escrever um programa que recebe 04 notas do aluno, calcula a média e imprimi aprovado
#recuperação ou reprovado

#regras

#utilizar while
#nota >= 7 aprovado
#nota >=5 recuperação
#nota <= 4 reprovado
bim = 1
nota = 0
soma = 0

try:

  while bim <= 4:
    nota = float(input(f"Digite sua nota do {bim}: "))
    bim +=1
    soma = soma + nota

  if (soma/4)>= 7:
    print("Parabens vc não é vagabundo e passou!!!")

  elif (soma/4)>= 5:
    print("droga vc ficou de recuperação T-T")

  else:
    print("seu vagabundo foi reprovado")

except:
  print("Digite apenas numeros")




#try = tentar
#except = "exceção"
```

```python
#calculadora simples

opcao = 0

while opcao != 4:

  print("opção 1 - seja bem vindo")
  print("opção 2 - soma")
  print("opção 3 - subtração")
  print("opção 4 - sair")
  opcao += 1
  opcao = int(input("Digite a opção desejada: "))
if opcao == 1:
  print("seja bem-vindos")

elif opcao == 2:
  a = float(input("Digite o numero: "))
  b = float(input("Digite o numero: "))
  print(f"A soma de {a} com {b} ficou = {a+b}")

elif opcao == 3:
  a = float(input("Digite o numero: "))
  b = float(input("Digite o numero: "))
  print(f"A subtração de {a} com {b} ficou = {a-b}")

else:
  print("Até uma proxima")
```
