```python
# 1° o usuario digita um numero e o programa escreve do 0 ao numero escolhido

num = int(input("digite o valor desejado: "))

for i in range(num + 1):
  print(i)
```

```python
# 2° contagem regressiva, o usuario digita um numero e o programa imprimi o num até o 0
num = int(input("digite o valor desejado: "))

for i in range(num,-1,-1):
  print(i)
```

```python
# 3° Somente imprimir os pares
num = int(input("digite o valor desejado: "))

for i in range(0,num + 1, 2):
  print(i)
```

```python
# 4°Somente imprimi os impares
num = int(input("digite o valor desejado: "))

for i in range(1,num + 1, 2):
  print(i)
```
