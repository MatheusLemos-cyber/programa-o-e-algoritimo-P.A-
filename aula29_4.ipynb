```python
# escrever um programa queo usuario digita uma palavra e o programa
# verifica se a palavra é uma palindroma e imprimir

palavra = str (input("Digite sua frase, e descopriremos se é palindroma: "))
invertida = ""

for i in (palavra):
    invertida = i + invertida

if invertida == palavra:
  print("a palavra escolhida é igual de frente ao contrario")
  print(f"{palavra} {invertida}")

else:
  print("esta palavra não é uma palindroma, fica diferente de frente para tras")
  print(f"{palavra} {invertida}")
```
