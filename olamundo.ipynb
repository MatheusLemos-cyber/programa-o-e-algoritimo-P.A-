### Criando uma Interface Simples com `ipywidgets`

`ipywidgets` é uma biblioteca que permite criar elementos de interface de usuário (UI) diretamente em notebooks Jupyter e Colab. Vamos criar um slider para escolher um número e um botão para exibir uma mensagem.

```python
import ipywidgets as widgets
from IPython.display import display, clear_output

# Criar um slider para escolher um número
slider = widgets.IntSlider(
    value=5,
    min=0,
    max=10,
    step=1,
    description='Escolha um número:',
    disabled=False,
    continuous_update=False,
    orientation='vertical',
    readout=True,
    readout_format='d'
)

# Criar um botão
button = widgets.Button(description="Clique-me!")

# Criar uma área de saída para exibir os resultados
output = widgets.Output()

# Definir o que acontece quando o botão é clicado
def on_button_clicked(b):
    with output:
        clear_output() # Limpa a saída anterior
        print(f"Você escolheu o número: {slider.value}")

# Conectar o botão à função
button.on_click(on_button_clicked)

# Exibir os widgets
display(slider, button, output)
```

#### Como funciona:

1.  **`IntSlider`**: Cria um controle deslizante para selecionar um valor inteiro.
2.  **`Button`**: Cria um botão clicável.
3.  **`Output`**: É uma área onde a saída (como `print` ou visualizações) pode ser exibida interativamente.
4.  **`on_button_clicked`**: Esta função é executada toda vez que o botão é clicado. Ela limpa a saída anterior e imprime o valor atual do slider.
5.  **`button.on_click(on_button_clicked)`**: Conecta a função `on_button_clicked` ao evento de clique do botão.
6.  **`display(slider, button, output)`**: Renderiza os widgets no notebook.

Você pode explorar muitos outros tipos de widgets (caixas de texto, checkboxes, seletores de data, etc.) e combiná-los para criar interfaces mais complexas!

### Criando uma Roleta de Números Simples

```python
import ipywidgets as widgets
from IPython.display import display, clear_output
import random

# Criar um widget para o usuário inserir sua aposta de número
guess_input = widgets.IntText(
    value=5,
    description='Seu Número (1-30):',
    disabled=False,
    min=1,
    max=30
)

# Criar um botão para girar a roleta
spin_button = widgets.Button(description="Girar Roleta!")

# Criar uma área de saída para exibir os resultados
output_roulette = widgets.Output()

# Definir o que acontece quando o botão é clicado
def on_spin_button_clicked(b):
    with output_roulette:
        clear_output() # Limpa a saída anterior

        user_guess = guess_input.value
        roulette_number = random.randint(1, 10) # Gera um número aleatório entre 1 e 10

        print(f"Você apostou no número: {user_guess}")
        print(f"O número da roleta foi: {roulette_number}")

        if user_guess == roulette_number:
            print("Parabéns! Você GANHOU!")
        else:
            print("Que pena! Você PERDEU.")

# Conectar o botão à função
spin_button.on_click(on_spin_button_clicked)

# Exibir os widgets
display(guess_input, spin_button, output_roulette)
```

#### Como funciona esta roleta:

1.  **`IntText`**: Onde você digita o número que deseja apostar (entre 1 e 10).
2.  **`Button`**: O botão "Girar Roleta!" que inicia a rodada.
3.  **`random.randint(1, 10)`**: Esta função gera um número inteiro aleatório entre 1 e 10 (inclusive).
4.  **`on_spin_button_clicked`**: Esta função é executada ao clicar no botão:
    *   Pega o número que você inseriu.
    *   Gera o número aleatório da roleta.
    *   Compara os dois números.
    *   Informa se você ganhou ou perdeu na área de saída (`output_roulette`).

Você pode adaptar o intervalo de números, adicionar mais condições de ganho/perda (por exemplo, números pares/ímpares, maior/menor), e até mesmo incluir um sistema de pontuação ou saldo para tornar o jogo mais complexo!
