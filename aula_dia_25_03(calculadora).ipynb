```python
# escrever um programa que recebe 04 notas do aluno, calcula a media e mostra se o aluno foi aprovado, recuperação ou reprovado

#entrada
print ("digite suas notas AQUI, valores de 01 a 10:")
nota = float (input("digite a nota do primeiro bimestre: "))
nota1 = float (input("digite a nota do segundo bimestre: "))
nota2 = float (input("digite a nota do terceiro bimestre: "))
nota3 = float (input("digite a nota do quarto bimestre: "))

#processamento

resultado = (nota + nota1 + nota2 + nota3) / 4

if resultado >= 7:
  print (f"a media das suas notas foi {resultado} você foi aprovado:")

elif resultado >= 5:
  print (f"a media das suas notas foi {resultado} você está de recuperação")

else:
  print (f" a media das suas notas foi {resultado}, você foi reprovado")


```
