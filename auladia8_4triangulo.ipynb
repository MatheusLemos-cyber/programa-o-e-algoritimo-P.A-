```python
#fazer um codigo que identifique se o numero é parou impar
# usando  "=" épara uma incognata "x" receber um valor
#usando "==" éliteralmente o igual da matematica
#entrada

a = int (input("Digite o numero que deseja descobrir se é par ou impar: "))

if a % 2 == 0:
   print (f"O numero {a}, é um numero par")

else:
  print (f"O numero {a}, é um numero impar")
```

```python
#Escrever um programa que o Usuario diga seu nome e suas 4  notas referentes do bimestre, calcule a média e diga se ele foi:
#aprovado, recuperação ou reprovado

#input abre um campo de digitação
#str = escrita
#int = numeros inteiras
#float = numeros quebrado (com virgula)
#segue a mesma regra da matematica "primeiro () e etc..."
#o comando "or" é literalmente "ou"

#nota 7 ou > é aprovado
#nota 5 ou < que 7 é recuperação
#nota 4 ou < é reprovado
#Entrada

nome = str (input("Digite seu nome: "))

nota1 = float (input("Digite a sua nota do 1° bimestre: "))
if nota1 > 10:
  print(f"digite uma nota valida dentre 0 à 10")

nota2 = float (input("Digite a sua nota do 2° bimestre: "))
if nota2 > 10:
  print(f"digite uma nota valida dentre o à 10")

nota3 = float (input("Digite a sua nota do 3° bimestre: "))
if nota3 > 10:
  print(f"digite uma nota valida dentre 0 à 10")

nota4 = float (input("Digite a sua nota do 4° bimestre: "))
if nota4 > 10:
  print(f"digite uma nota valida dentre 0 à 10")

media = (nota1 + nota2 + nota3 + nota4) / 4

if media >= 7:
  print(f"{nome} está aprovado, com uma media de {media}: ")

elif media >= 5:
  print(f"{nome} está em recuperação, com uma media de {media}: ")

else:
  print(f"{nome} está reprovado, com uma media de {media}:")
```

```python
#escrever um programa que recebe os valores de 03 lados e verifique se é um triangulgo:
#valido (a somo de 2 lados sempre tem que ser maior que o 3 lado)
# classificar tipos de triangulos
# * Equilatero= Todos os lados iguais
# * Isoceles= possui dois lados iguais
# * escaleno= Todos os lados diferentes

#se o comando estiver na linha de baixo, com espaço, seria igual uma continuidade
#do comando de cima

#processamento

lado1 = float (input("Digite o valor do lado A: "))
lado2 = float (input("Digite o valor do lado B: "))
lado3 = float (input("Digite o valor do lado C: "))

if (lado1 + lado2 > lado3) and (lado1 + lado3 > lado2) and (lado2 + lado3 > lado1):
  print(f"Este triangulo é valido.")
else:
  print(f"este triangulo não é valido")
```
